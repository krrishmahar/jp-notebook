### 2 ways to import data from sklearn.datasets

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1)
mnist.keys()
```

```python
X,y = fetch_openml('mnist_784', version=1, return_X_y=True, as_frame=False)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
some_digit = X[0]
some_digit_image = some_digit.reshape(28, 28)
plt.imshow(some_digit_image, cmap = mpl.cm.binary, interpolation="nearest")
plt.axis("off")
plt.show()
```

## Splitting into Training and test set

```python
# Casting values as integers
y = y.astype(np.uint8)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
y_train5 = (y_train == 5)
y_test5 = (y_test == 5)
```

### Stochastic Gradient Descent is sensitive to feature scaling, so it is highly recommended to scale your data.
### If your attributes have an intrinsic scale (e.g. Image from 0 - 255, word frequencies or indicator features) scaling is not needed. 

#### 2 models below

```python
from sklearn.pipeline import make_pipeline
est = make_pipeline(SGDClassifier())
est.fit(X_train, y_train5)
est.predict(X_test)
```

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train5)
```

```python
sgd_clf.predict(some_digit.reshape(1,-1))
```
